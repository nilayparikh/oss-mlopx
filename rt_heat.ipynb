```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
file_path = '.data/MKT_DATA_EVAL.csv'

# Read the CSV file and load it into a DataFrame
df = pd.read_csv(file_path)

plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x="threads", y="runtime_secs", cmap="Reds", fill=True)

plt.title('Runtime Heat Area')
plt.xlabel('Number of Threads')
plt.ylabel('Runtime (seconds)')

plt.show()  
```

```python
import matplotlib.ticker as ticker

file_path = '.data/RISK_MODEL.csv'

# Read the CSV file and load it into a DataFrame
df = pd.read_csv(file_path)

plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x="threads", y="runtime_secs", cmap="Reds", fill=True)

plt.title('Runtime Heat Area')
plt.xlabel('Number of Threads')
plt.ylabel('Runtime (seconds)')

# Modify x-axis ticks to only show whole numbers
ax = plt.gca()
ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

plt.show()
```
